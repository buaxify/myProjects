```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
close = pd.read_csv(r"C:\Users\Emilie Berger\Downloads\sp500_close.csv", index_col=0, parse_dates=True)
goog = close[['GOOG']].copy().dropna()
goog.head()
```

```python
goog.plot()
```

```python
#google log_return by calling pct_change function by adding 1 and taking the log
goog_ret = np.log(goog.pct_change(1) + 1)
goog_ret.plot()
```

```python
#test moving average in code
goog['SMA - 10'] = goog['GOOG'].rolling(10).mean()
goog.head(20)
```

```python
type(goog['GOOG'].rolling(10))
```

```python
goog.plot(figsize=(10, 5))

#SMA - 10 smoother version of original timeseries
```

```python
goog['SMA - 50'] = goog['GOOG'].rolling(50).mean()
goog.plot(figsize=(10, 5))

#even smoother, effect become more and more pronouced as window size gets larger
```

```python
# multi-dimensional time series
goog_aapl = close[['GOOG', 'AAPL']].copy().dropna()
cov = goog_aapl.rolling(50).cov()
cov
```

```python
cov.loc['2018-02-07'].to_numpy()
```

```python
goog_aapl_ret = np.log(1 + goog_aapl.pct_change(1))
goog_aapl_ret.head()
```

```python
goog_aapl_ret['GOOG-SMA-50'] = goog_aapl_ret['GOOG'].rolling(50).mean()
goog_aapl_ret['AAPL-SMA-50'] = goog_aapl_ret['AAPL'].rolling(50).mean()
goog_aapl_ret.plot(figsize=(10, 5))
```
