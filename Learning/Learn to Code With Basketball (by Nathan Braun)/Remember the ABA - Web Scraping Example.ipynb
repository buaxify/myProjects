## Import Libraries

```python
from bs4 import BeautifulSoup as Soup
import pandas as pd
import requests
from pandas import DataFrame
```

## URL visit --> HTML store

```python
response = requests.get('http://www.remembertheaba.com/ABAStatistics/ABATop20.html')
```

```python
print(response.text)
```

## Let's parse it

```python
aba_soup = Soup(response.text)
aba_soup
```

## Let's find all the table tags in this html

```python
tables = aba_soup.find_all('table')
tables
```

## How many tables we got back

```python
len(tables)
```

## let's pick 1

```python
aba_table = tables[0]
```

```python
rows = aba_table.find_all('tr')
rows
```

```python
first_data_row = rows[2]
first_data_row
```

```python
first_data_row.find_all('td')
```

```python
[str(x.string) for x in first_data_row.find_all('td')]
```

```python
def parse_row(row):
    return [str(x.string) for x in row.find_all('td')]

list_of_parsed_rows = [parse_row(row) for row in rows[2:]]
list_of_parsed_rows
```

```python
df = DataFrame(list_of_parsed_rows)
df.head()
```

```python
df.columns = [str(x.string).lower() for x in rows[1].find_all('th')]
df.columns
```

```python
str_cols = ['name', 'years']
float_cols = ['fg%', '3p%', 'ft%', 'ppg']

int_cols = [x for x in df.columns if not ((x in str_cols) or 
                                          (x in float_cols))]
int_cols
```

```python
df[float_cols] = df[float_cols].astype(float)
```

```python
# df[int_cols] = df[int_cols].astype(int) -->
# invalid literal for int() with base 10: '-\xa0'

# df[int_cols] = pd.to_numeric(df[int_cols], errors='coerce') -->
# arg must be a list, tuple, 1-d array, or Series

for int_col in int_cols:
    df[int_col] = pd.to_numeric(df[int_col], errors='coerce')
    
df.head()
```
