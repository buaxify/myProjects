# Open Excel File

```python
# Import modules
import openpyxl 
import pandas as pd


File = 'Y:\IT\GCG Online - Portal Pixels\VBA\SmallInvoice\Invoices\Invoices Reporting 2022_VBA.xlsm'
File2 = 'Y:\IT\GCG Online - Portal Pixels\VBA\SmallInvoice\Invoices\Invoices Reporting 2022_VBA - 2.xlsm'

# Read Excel file as Pandas DataFrame
df = pd.read_excel(File)

# Open an Excel workbook
workbook = openpyxl.load_workbook(File, False, True)

# Create a `Worksheet` object 
worksheet = workbook['Factures']

# List of indices corresponding to all hidden rows
hidden_rows_idx = [
    row - 2
    for row, dimension in worksheet.row_dimensions.items() 
    if dimension.hidden
]

# f = open(File,"w+")
# f.close()

#workbook.close(File)




df.drop(hidden_rows_idx, axis=0, inplace=True)
lastRow = len(df.index)
df
```
