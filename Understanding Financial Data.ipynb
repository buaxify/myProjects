```python
dates = pd.date_range(df['date'].min(), df['date'].max())
dates
```

```python
close_prices = pd.DataFrame(index=dates)
```

```python
close_prices.head()
```

```python
symbols = df['Name'].unique()
```

```python
df2 = pd.DataFrame(data=ibm['close'].to_numpy(), index=ibm['date'], \
                   columns=['IBM'])
```

```python
df2.head()
```

```python
#Let's now fill in our close prices dataframes
for symbol in symbols:
    df_sym = df[df['Name'] == symbol]
    df_tmp = pd.DataFrame(data=df_sym['close'].to_numpy(), \
                          index=df_sym['date'], columns=[symbol])
    close_prices = close_prices.join(df_tmp) #left-join by default
```

```python
#Now it has dates, but also missing values
close_prices['IBM'].plot()
```

```python
# drop df rows with nan (most likely weekends, holidays...)
close_prices.dropna(axis=0, how='all', inplace=True)
```

```python
close_prices['IBM'].plot()
```

```python
close_prices.plot(legend=None, figsize=(10, 10))
```

```python
close_prices.isna().sum() #how many missing values per row
```

```python
close_prices.isna().sum().sum() #how many missing values in total
```

```python
close_prices.iloc[0, :].isna().sum() #how many missing values in 1st row
```

```python
close_prices.fillna(method= 'ffill', inplace=True) #Forward Fill
```

```python
close_prices.isna().sum().sum() #still values missing?
```

```python
close_prices.fillna(method= 'bfill', inplace=True) #Backward Fill
```

```python
close_prices.isna().sum().sum() #still values missing?
```

```python
close_prices.plot(legend=False, figsize=(10, 10))
```

```python
#relative stock price overtime --> make cumulative returns more visible --> compare growth
close_prices_normalized = close_prices / close_prices.iloc[0] #close price divided by initial close price
```

```python
close_prices_normalized.plot(legend=False, figsize=(10, 10))
```
